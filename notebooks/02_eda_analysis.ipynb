# UK Cost of Living Analysis - EDA

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
housing = pd.read_csv('../data/processed/housing_merged_clean.csv')
cpi = pd.read_csv('../data/processed/cpi_reference_tables_clean.csv')
master = pd.read_csv('../data/processed/master_housing_cpi_dataset.csv')

housing['date'] = pd.to_datetime(housing['date'], errors='coerce')
cpi['date'] = pd.to_datetime(cpi['date'], errors='coerce')
master['date'] = pd.to_datetime(master['date'], errors='coerce')

print(housing.shape, cpi.shape, master.shape)
```

```
plt.figure()
plt.plot(cpi['date'], cpi['cpi_index'])
plt.title('UK CPI Trend')
plt.show()
```

```
plt.figure()
plt.plot(housing['date'], housing['e09000001_average_price'])
plt.title('London Housing Prices')
plt.show()
```

```
plt.figure()
plt.plot(master['date'], master['cpi_index'], label='CPI')
plt.plot(master['date'], master['e09000001_average_price'], label='Housing')
plt.legend()
plt.show()
```
